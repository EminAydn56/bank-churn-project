```python
import pandas as pd

df = pd.read_csv("churn.csv")
df.head()
```

```python
df['churn'].value_counts()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(x='churn', data=df)
plt.title("Customer Churn Distribution")
plt.show()
```

## Insight 1
Most customers are staying with the bank, but a noticeable number have churned.

```python
sns.boxplot(x='churn', y='age', data=df)
plt.title("Age vs Churn")
plt.show()
```

## Insight 2
Older customers tend to leave the bank more frequently.

```python
sns.boxplot(x='churn', y='estimated_salary', data=df)
plt.title("Salary vs Churn")
plt.show()
```

## Insight 3
Salary does not have a strong impact on customer churn.

```python
sns.countplot(x='active_member', hue='churn', data=df)
plt.title("Active Member vs Churn")
plt.show()
```

## Insight 4
Inactive customers are more likely to leave the bank.
